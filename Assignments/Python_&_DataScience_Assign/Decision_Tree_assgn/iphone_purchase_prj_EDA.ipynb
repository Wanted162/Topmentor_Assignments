```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/iphone_purchase_records.csv")
df.head()
```

```python
##summary stats
df.describe()
```

```python
##data type of each column
df.dtypes
```

```python
##check if there are any null values
df.isnull().sum()
```

```python
##visualize the data by distributing the age
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.savefig('age_distribution.png')
plt.show()
```

```python
##visualize the data by distributing the salary
plt.figure(figsize=(10,6))
sns.histplot(df['Salary'], kde=True)
plt.title("Salary Distribution")
plt.savefig('salary_distribution.png')
plt.show()
```

```python
## purchase count by gender
plt.figure(figsize=(10,6))
sns.countplot(x='Gender', hue='Purchase Iphone', data=df)
plt.title("purchase count by gender")
plt.savefig('purchase_count_gender.png')
plt.show()
```

```python
##age vs salary
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='Salary', hue='Purchase Iphone', data=df)
plt.title("Age vs Salary")
plt.savefig('Age_vs_salary.png')
plt.show()
```

```python
##save the cleaned dataset
df.to_csv('cleaned_iphone_purchase.csv', index=False)
```
